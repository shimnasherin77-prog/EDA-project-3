```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
order=pd.read_csv("furniture.csv")
order.head(50)
```

```python
sns.barplot(x='inventory',y='material',data = order,orient='h')
```

```python
sns.barplot(x='inventory',y='category',data = order,orient='h')
```

```python
season_sales = order.groupby("season")["sales"].sum()
print(season_sales)
plt.figure(figsize=(7,7))
plt.pie(season_sales,labels=season_sales.index,
         autopct='%1.1f%%',startangle=90,shadow=True)
plt.title("Sales Distribution by Season")
plt.show()
```

```python
sns.barplot(x='category',y='color',data = order,orient='h')
```

```python

plt.figure(figsize=(10,6))
sns.histplot(order['color'],bins = 10)
```

```python
sns.barplot(x="location", y="revenue", data=order, estimator=sum, palette="viridis")
```

```python
sns.barplot(x='location', y='delivery_days', data=order, estimator='mean', palette='viridis')
```

```python
sns.jointplot(x='cost',y='sales',data=order,kind='kde',fill='True')
```

```python
sns.jointplot(x='cost',y='sales',data=order,kind='hex')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='season', y='revenue', data=order, palette='coolwarm')
```

```python
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.stripplot(x='material', y='cost', data=order, jitter=True, palette='Set3', size=6)
plt.legend(loc=2)
```
